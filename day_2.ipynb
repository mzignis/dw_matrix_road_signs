```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
```

```
HOME = '/content/drive/My Drive/Colab Notebooks/matrix/dw_matrix_road_signs'
%cd $HOME
```

```
train_db = pd.read_pickle('data/train.p')
train_db.keys()
```

```
X_train, y_train = train_db['features'], train_db['labels']
X_train.shape, y_train.shape
```

```
plt.imshow(X_train[0])
```

```
signs = pd.read_csv('data/dw_signnames.csv')
dict_signs = signs.to_dict()['b']
dict_signs[30]
```

```
for sign_id in dict_signs.keys():
    selected_signs = X_train[np.where(y_train == sign_id)]
    plt.figure(figsize=(15, 5))
    for idx in range(9):
        plt.subplot(f'19{idx}')
        plt.imshow(selected_signs[idx])
        plt.axis('off')
```

```
cnt = Counter(y_train).most_common()
idx_labels, cnt_labels = zip(*cnt)
ids = range(len(idx_labels))
```

```
plt.figure(figsize=(15, 5))
plt.bar(ids, cnt_labels)
plt.xlabel('Znaki')
plt.xticks(ids, [dict_signs[idx_labels[id_]] for id_ in idx_labels], rotation='vertical')
plt.ylabel('Ilosc znakow w bazie')
```
